```python
# @title Click â–º to run program!

import shutil
import os
import zipfile

def copy_and_rename_file(source_path, destination_directory, new_file_name):
    destination_path = os.path.join(destination_directory, new_file_name)

    shutil.copy(source_path, destination_path)
    print(f"File copied and renamed to '{new_file_name}'.")

def zip_directory(source_dir, output_zip_file):
    """
    Creates a zip archive of the specified directory.

    Args:
        source_dir (str): The path to the directory to be zipped.
        output_zip_file (str): The path and filename for the output zip archive.
    """
    with zipfile.ZipFile(output_zip_file, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(source_dir):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, source_dir)
                zipf.write(file_path, arcname)

# Inputs
filenames = ["Technical Theatre Magnet Program", "Theatre Magnet Program", "Band Magnet Program", "Cinematic Arts Magnet Program",
             "Costume Design Magnet Program", "Dance Magnet Program", "Mariachi_ Music Magnet Program", "Orchestra Magnet Program",
             "Visual Arts Magnet Program", "Vocal Magnet Program"]
destination_folder = "addi_cafe_videos"
zip_folder = destination_folder + ".zip"
video_file = ".mp4"


source_file = input(">> Enter source file name: ")
file_count = input(">> Enter file count: ")
print("")
print("Duplicating file: " + source_file[:-4])
print("File count: " + file_count)
print("File type: " + video_file)
continue_program = input(">> Continue [y/n]? ")

if continue_program == "n":
  exit()

os.makedirs(destination_folder, exist_ok=True)

print("")
print("Duplication in progress...")
for i in filenames:
  renamed_file = i + " " + file_count + video_file
  copy_and_rename_file(source_file, destination_folder, renamed_file)
print("")
print("Zipping folder...")
zip_directory(destination_folder, zip_folder)
print("")
print("Removing source folder...")
shutil.rmtree(destination_folder)

print("\nEnd of program...")
```
